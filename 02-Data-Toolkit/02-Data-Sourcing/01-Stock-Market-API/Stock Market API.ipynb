```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib
```

```python
import requests

url = "https://iex.lewagon.com/stable/stock/aapl/chart/3m"
api_data = requests.get(url).json()
```

```python
apple_df = pd.DataFrame(api_data)
```

```python
#Convert the date to a datetime object
apple_df['date'] = pd.to_datetime(apple_df['date'], format = "%Y-%m-%d")
apple_df
```

```python
#Set the date as the index
apple_df.set_index(apple_df['date'], inplace = True)
apple_df
```

```python
#Plot values in close column
apple_df[['open', 'close', 'high', 'low']].plot(figsize = (15,10))
```
